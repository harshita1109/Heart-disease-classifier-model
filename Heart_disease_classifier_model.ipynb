**COLUMN DESCRIPTION**
1. Age - age of patient

2. Sex - gender of patient (1-male,0-female)

3. CP - type of chest pain(1-typical angina , 2- atypical angina , 3 - non-anginal pain , 4- asymptomatic)

4. Tresttbps - the resting blood pressure of the patient in mm hg.

5. Chol - serum of cholestrol level of the patient in mg or dl

6. FBS - fasting blood sugar level of the patient (1-high,0-low)

7. Restecg - Rest electro cardio graphic results of patient (0-normal 1- ST-T wave abnormality , 2= left venticular hypertrophy)

8. thelach - maximum heart rate achievved by the patient during exercise.

9. oldpeak - the ST depression induced by exercise relative to rest.
SLOPE of ST segment during peak exercise.(1- upslopping , 2- flat , 3- dowmslopping)

10. ca- number of vessels coloured by fluoroscopy(0-3).

11. thal - type of thallium scan that is performed on the patient (1- fixed defect, 2- reversible defect, 3- normal)

12. Target - Presence of fat disease-(0-no disease, 1- disease present)

```python
from sklearn.model_selection import train_test_split                 # to split the data into training and testing
from sklearn.metrics import accuracy_score,confusion_matrix          # to evaluate the model performance
from sklearn.tree import DecisionTreeClassifier                      # to load the model library
```

```python
# DIVIDING THE DATA INTO TRAINING AND TESTING
X=df.drop('target',axis=1)
X
```

```python
y=df['target']
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
accuracy_score(y_pred,y_test)*100
```

```python
# GETTING CONFUSION MATRIX
confusion_matrix(y_pred,y_test)
```
